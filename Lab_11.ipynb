# Travelling Salesman Problem using Dynamic Programming.
Problem Description:
Given a set of cities and distances, find the shortest possible route that visits each city once and returns to the starting city.

#Theory
TSP uses dynamic programming with bit masking. Store results of visited subsets to reduce repeated work.

## Algorithm
1. Start from source city
2. Mark visited cities using bit mask
3. Try all unvisited cities
4. Choose minimum cost path
5. Return to source city

```python
graph = [
[0, 10, 15, 20],
[10, 0, 35, 25],
[15, 35, 0, 30],
[20, 25, 30, 0]
]

print("Minimum travelling cost:", tsp(graph))
```

```python
import numpy as np

def tsp(graph):
    n = len(graph)
    # dp[city][mask] stores the minimum cost to visit remaining cities starting from 'city' with 'mask' visited
    dp = [[-1]*(1<<n) for _ in range(n)]
    # path_store[city][mask] stores the next city in the optimal path from 'city' with 'mask' visited
    path_store = [[-1]*(1<<n) for _ in range(n)]

    def visit(city, mask):
        if mask == (1<<n) - 1: # All cities visited, return to start (city 0)
            return graph[city][0]

        if dp[city][mask] != -1:
            return dp[city][mask]

        ans = float('inf')
        best_next_city = -1

        for next_city_node in range(n):
            # Check if next_city_node is not visited and there's a path from current city
            if not (mask & (1<<next_city_node)) and graph[city][next_city_node] != 0: # Ensure valid path (distance > 0 for non-self)
                new_cost = graph[city][next_city_node] + visit(next_city_node, mask | (1<<next_city_node))
                if new_cost < ans:
                    ans = new_cost
                    best_next_city = next_city_node

        dp[city][mask] = ans
        path_store[city][mask] = best_next_city
        return ans

    min_cost = visit(0, 1) # Start from city 0, with 0 visited (mask = 1)

    # Reconstruct the path
    path = [0] # Start from city 0
    current_city = 0
    current_mask = 1 # Mark city 0 as visited

    while current_mask != (1<<n) - 1:
        next_chosen_city = path_store[current_city][current_mask]
        if next_chosen_city == -1: # Should not happen if path_store is correctly populated
            print("Error: Could not reconstruct full path. Path_store might be incomplete.")
            break
        path.append(next_chosen_city)
        current_city = next_chosen_city
        current_mask |= (1 << current_city) # Mark current_city as visited

    path.append(0) # Return to the starting city

    return min_cost, path
```

### Part 2: Take Graph Input from User
This function allows you to manually input the distance matrix for your TSP problem. After running the function, you'll be prompted to enter the number of cities and then the distances row by row. Uncomment the example usage to run it.

```python
def get_user_graph():
    while True:
        try:
            n = int(input("Enter the number of cities (n): "))
            if n <= 1:
                print("Number of cities must be greater than 1.")
                continue
            break
        except ValueError:
            print("Invalid input. Please enter an integer.")

    graph = []
    print(f"Enter the {n}x{n} distance matrix row by row. Use spaces to separate values.")
    print("Example: For a 3x3 graph, first row might be '0 10 15'")

    for i in range(n):
        while True:
            try:
                row_str = input(f"Enter row {i+1} (n values): ")
                row = list(map(int, row_str.split()))
                if len(row) != n:
                    print(f"Incorrect number of values. Expected {n}, got {len(row)}.")
                    continue
                graph.append(row)
                break
            except ValueError:
                print("Invalid input. Please enter integers separated by spaces.")

    return graph

# Example of using user input (uncomment the lines below to run)
# print("\n--- User Input Graph ---")
# user_graph = get_user_graph()
# if user_graph:
#     cost, path = tsp(user_graph)
#     print("Minimum travelling cost (user graph):", cost)
#     print("Optimal path (user graph):", path)
```
